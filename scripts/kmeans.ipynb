```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.cluster import KMeans
import preprocessing
import sklearn.metrics
```

```python
data=pd.read_csv('../src/jain_out.txt')

data.columns=['x','y','classe']
data.drop(['classe'],axis=1)
x=data.iloc[:,0]
y=data.iloc[:,1]
labelsjain=data.iloc[:,2]
plt.xlabel("x")
plt.ylabel("y")
plt.title("Clustering Jain.txt Fournies")
plt.scatter(x,y,c=labelsjain)
plt.show()
```

```python
data=pd.read_csv('../src/jain_out.txt')

def kmeans(data,nbclusters):
    kmeans=KMeans(n_clusters=nbclusters,init="k-means++", n_init=2)
    kmeans.fit(data)
    kmeans.predict(data)
    reskmeans=kmeans.transform(data)
    return(kmeans.cluster_centers_,kmeans.labels_,reskmeans)

labelskmeansjain=kmeans(data,2)[1]
x=data.iloc[:,0]
y=data.iloc[:,1]
plt.xlabel("x")
plt.ylabel("y")
plt.title("Données Jain.txt Traitées par Kmeans")
plt.scatter(x,y,c=labelskmeansjain)
plt.show()

#Comparaison avec les données :
rand = sklearn.metrics.adjusted_rand_score(labelsjain,labelskmeansjain)
print("Indice de Rand Ajusté = ",rand)
print("Choix de 2 Clusters")
```
