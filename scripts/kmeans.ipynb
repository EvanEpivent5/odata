```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.cluster import KMeans
import preprocessing
import sklearn.metrics
```

```python
def jain():
    data=pd.read_csv('../src/jain_out.txt')

    data.columns=['x','y','classe']
    data.drop(['classe'],axis=1)
    x=data.iloc[:,0]
    y=data.iloc[:,1]
    labelsjain=data.iloc[:,2]
    plt.xlabel("x")
    plt.ylabel("y")
    plt.title("Clustering Jain.txt Fournies")
    plt.scatter(x,y,c=labelsjain)
    plt.show()
jain()
```

```python
def kmeans(data,n_clusters,init,n_init):
    kmeans=KMeans(n_clusters,init,n_init)
    kmeans.fit(data)
    kmeans.predict(data)
    reskmeans=kmeans.transform(data)

    labelskmeans=kmeans.labels_
    x=data.iloc[:,0]
    y=data.iloc[:,1]
    plt.xlabel("x")
    plt.ylabel("y")
    plt.title("Donnes .txt Traites par Kmeans")
    plt.scatter(x,y,c=labelskmeans)
    plt.show()

    #Comparaison avec les donnes :
    rand = sklearn.metrics.adjusted_rand_score(data.iloc[:,2],labelskmeans)
    print("Indice de Rand Ajust = ",rand)
    print("Choix de ",n_clusters," Clusters")

jain=pd.read_csv('../src/jain_out.txt')
kmeans(jain,2,"k-means++",2)
```
